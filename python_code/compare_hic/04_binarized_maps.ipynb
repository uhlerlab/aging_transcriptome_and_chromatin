```python
def get_stats(young, old):
    diff = young - old
    return np.sum(np.sum(young))
    
get_stats(IMR90, old_fibroblasts)
```

# 3. Difference map for last PCST network loci (design 2)

```python
# Plot binarized maps
IMR90_net = pd.read_csv(hic_dir+'processed_hic_data_IMR90/binarized_maps/last_PCST_subset.csv', index_col = 0)
old_fibroblasts_net = pd.read_csv(hic_dir+'processed_hic_data_old_fibroblasts/binarized_maps/last_PCST_subset.csv', index_col = 0)

hic.plot_binarized_maps(IMR90_net, old_fibroblasts_net)
plt.savefig(fig_dir + 'binarized_hic_last_PCST.png')
```

```python
# Plot differences
diff_net = IMR90_net - old_fibroblasts_net
diff_net.to_csv(hic_dir + 'difference_maps/diff_last_PCST_loci.csv')

plt.figure()
sns.heatmap(diff_net, cmap = "vlag", xticklabels=False, yticklabels=False)
plt.title("Difference map for last PCST network")
plt.ylabel("")
plt.savefig(fig_dir + 'difference_map_last_PCST.png')
plt.show()
```

```python
# Plot clustered differences
plt.figure()
sns.clustermap(diff_net.loc[(diff_net!=0).any(axis=1), (diff_net!=0).any(axis=1)],
               method='ward',
               metric='euclidean',
               row_cluster=True, col_cluster=True,
               figsize=(6,6),
               xticklabels=False, yticklabels=False,
               cmap='vlag', cbar_pos=(1, 0.5, 0.01, .4),
               vmin=-1, vmax=1,
               dendrogram_ratio=(.1, .1))
plt.savefig(fig_dir+'clustered_diff_map_last_PCST.png')
plt.show()
```

```python
# Hierarchical clustering
# create dendrogram
corr = diff_net.loc[(diff_net!=0).any(axis=1), (diff_net!=0).any(axis=1)]#.corr()
dend = linkage(corr, method='ward', metric='euclidean')

plt.figure(figsize = (6, 8))
dendrogram(dend, orientation = "left", color_threshold = 24)
#plt.savefig(save_dir+'dendrogram_diff_last_PCST.png')
plt.show()
```

```python
clusters = fcluster(dend, 24, criterion = 'distance')
loci_clusters = pd.DataFrame({'locus': corr.index, 'cluster': clusters})
loci_clusters['chrom'] = loci_clusters['locus'].str.split('_').str[1]
#loci_clusters.to_csv(hic_dir + 'difference_maps/clusters_last_PCST.csv', index = False)
```

```python
unique, counts = np.unique(clusters, return_counts=True)
print(unique, counts)
#1 = orange, 2 = green, 3 = red, 4 = purple
```

```python
# Output as a graph
diff_long = pd.melt(diff_net.reset_index(), id_vars='chr1')
diff_long = diff_long[diff_long['value'] != 0]
diff_long.columns = ['loc1', 'loc2', 'diff']
network = nx.from_pandas_edgelist(diff_long, 'loc1', 'loc2', 'diff')

# add clusters as node attributes
clusters_dict = {loci_clusters.loc[index, 'locus']: str(loci_clusters.loc[index, 'cluster']) for index in loci_clusters.index}
nx.set_node_attributes(network, clusters_dict, name='hierarchical-cluster')

#oi.output_networkx_graph_as_interactive_html(network, filename=fig_dir + "diff_net_last_PCST.html")
```

## Community detection: Louvain clusters

```python
partition = community.best_partition(network)
partition = {chrom: str(cluster) for chrom, cluster in partition.items()}
nx.set_node_attributes(network, partition, name='louvain-cluster')
oi.output_networkx_graph_as_interactive_html(network, filename=fig_dir + "diff_net_last_PCST.html")
```

```python
louvain_clusters = pd.DataFrame.from_dict(partition, orient='index').reset_index()
louvain_clusters.columns = ['locus', 'cluster']

# add corresponding gene(s) for each locus
net = pickle.load(open(pcst_dir + "network_allTFs_200_step_27-60_61-85_86-96" +  
                                            "_design_2.pickle", "rb"))
selected_genes = [gene[:-4] for gene in net.nodes()]
gene_loci = pd.read_csv(genome_dir + 'all_gene_loci.csv')
gene_loci = gene_loci[gene_loci['gene'].isin(selected_genes)]

louvain_clusters = louvain_clusters.merge(gene_loci, how = "left", on = 'locus')
louvain_clusters.to_csv(hic_dir + 'difference_maps/louvain_clusters_last_PCST.csv', index = False)
```
